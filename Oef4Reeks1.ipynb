```python
from sympy import *
init_printing()
x, y, z, u, v,r = symbols('x y z u v,r')
alpha1, beta, omega_21 = symbols('alpha1 beta omega_2_1')
x, y, z, alpha1, beta, omega_21
```

```python
# x,y,z,r uitrekennen 
EQ1 = Eq((-3-x)**2+(3-y)**2+(-4-z)**2 - r**2 )
EQ2 = Eq((1-x)**2+(1-y)**2+(z)**2-r**2)
EQ3 = Eq((3-x)**2+(1-y)**2+(-2-z)**2 -r**2)
EQ4 = Eq((-5-x)**2+(5-y)**2+(-4-z)**2 -r**2)
solve([EQ1,EQ2,EQ3,EQ4], (x, y, z, r))
# hiermee hebben we de vergelijking van de bol : (x-2)**2 + (y-10)**2 +(z+1)**2 = 83
# bij controle door de punten in te geven komen we 4 geldige vergelijkingen uit, en in geogebra klopt het ook.
# de straal is gelijk aan sqrt(83)
# het middelpunt is gelijk aan (2,10,-1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Middelpunt en straal van de bol
center = (2, 10, -1)
radius = np.sqrt(83)

# Maak een grid van punten voor de bol
u = np.linspace(0, 2 * np.pi, 100)  # Hoek in de xy-vlak
v = np.linspace(0, np.pi, 100)       # Hoek in de yz-vlak

# Parametriseer de bol
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Coördinaten van de punten A, B, C en D
points = {
    'A': (-3, 3, 4),
    'B': (3, 1, -2),
    'C': (1, 1, 0),
    'D': (-5, 5, 0)
}

# Maak een 3D-plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot de bol
ax.plot_surface(x, y, z, color='w', alpha=0.6)  # Kleur en transparantie

# Plot de punten A, B, C en D
for point_name, (px, py, pz) in points.items():
    ax.scatter(px, py, pz, label=point_name, s=100)  # s = grootte van de marker

# Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('3D Bol met middelpunt (2, 10, -1) en straal √83')

# Toon de legenda
ax.legend()

# Toon de plot
plt.show()
```
